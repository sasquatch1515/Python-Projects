## #6 High and Low Spenders

```python
df_ord_prod_merge['user_id'].value_counts(dropna=False)
```

```python
## Adding column for average item price
df_ord_prod_merge['average_price'] = df_ord_prod_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
df_ord_prod_merge[['user_id', 'average_price']].head(50)
```

```python
## Creating a Flag for high and low spenders
df_ord_prod_merge.loc[df_ord_prod_merge['average_price'] < 10, 'spending_flag'] = 'low spender'
df_ord_prod_merge.loc[df_ord_prod_merge['average_price'] >= 10, 'spending_flag'] = 'high spender'
```

```python
df_ord_prod_merge[['user_id', 'average_price', 'spending_flag']].head(100)
```

## #7 Flagging Order Frequency by Customer

```python
## Adding column for median days since prior order
df_ord_prod_merge['median_order_freq'] = df_ord_prod_merge.groupby(['user_id'])['days_since_last_order'].transform('median')
```

```python
df_ord_prod_merge.head(50)
```

```python
## Adding flags for order frequency
df_ord_prod_merge.loc[df_ord_prod_merge['median_order_freq'] > 20, 'freq_flag'] = 'non-frequent customer'
df_ord_prod_merge.loc[(df_ord_prod_merge['median_order_freq'] <= 20) & (df_ord_prod_merge['median_order_freq'] > 10), 'freq_flag'] = 'Regular customer'
df_ord_prod_merge.loc[df_ord_prod_merge['median_order_freq'] <= 10, 'freq_flag'] = 'frequent customer'
```

```python
df_ord_prod_merge['freq_flag'].value_counts(dropna = False)
```
